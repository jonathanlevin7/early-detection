```python
import json
import matplotlib.pyplot as plt
import numpy as np

with open("../outputs/misclassified_samples.json", "r") as f:
    misclassified_data = json.load(f)

for sample in misclassified_data[:10]: #view first 10.
    image = np.array(sample["image"]).transpose(1, 2, 0) #rearrange axes.
    prediction = sample["prediction"]
    label = sample["label"]

    plt.imshow(image)
    plt.title(f"Predicted: {prediction}, Actual: {label}")
    plt.show()
```
